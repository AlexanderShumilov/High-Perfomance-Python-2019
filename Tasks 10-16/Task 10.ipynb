```python
import numpy as np
import cupy as cp
import matplotlib.pyplot as plt
import matplotlib.animation as animation
%matplotlib inline

def Saxpy(x, y, a):
    return a * x + y

def Saxpy_cp(x, y, a):
    x = cp.asarray(x)
    y = cp.asarray(y)
    return a * x + y
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time
%matplotlib inline

time_cuda = []
time_numpy = []
a = 10
Range = range(100000, 1000000, 10)
time_start = 0
for i in Range:
    x = np.random.random(i)
    y = np.random.random(i)
    
    time_start = time.time()
    Saxpy(x, y, a)
    time_numpy.append(time.time() - time_start)
    
    x = cp.random.random(i)
    y = cp.random.random(i)    
    time_start = time.time()
    Saxpy_cp(x, y, a)
    time_cuda.append(time.time() - time_start)
```

```python
plt.figure(figsize=(11, 9))
plt.plot(Range, time_numpy, label='Numpy')
plt.plot(Range, time_cuda, label='Cuda')
plt.rcParams.update({'font.size': 20})
plt.legend(loc='best')
#plt.xscale('log')
plt.yscale('log')
plt.show()
```
