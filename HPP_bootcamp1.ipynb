```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Definition of bifurcation function

def bifurcation(x, r):
    return(r * (1 - x) * x)
```

```python
# Test for one particular r

Range = 101
r = 2.5
X = []
x = np.random.rand()

for n in range(Range):
    x = bifurcation(x, r)
    X.append(x)
plt.plot(range(Range), X, '.')
plt.xlabel('N')
plt.ylabel('x')
plt.show() 
```

```python
# As we can see, with r-value between 2 and 3 everything is diverges (so we kind of played with r)

plt.figure(figsize=(10,10))
Range = 101
r = 2.5
x = np.random.rand()

for i in range(9):
    X = []
    for n in range(Range):
        x = bifurcation(x, r)
        X.append(x)
    r = r + 0.1
    plt.plot(range(Range), X, '.', label = 'r = %.1f' % r)


plt.legend(loc='best')
plt.xlabel('N')
plt.ylabel('x')
plt.show() 
```

```python
# Here we can observe bifurcation with 200 elements of our X_n left
Range = 1001
r = np.linspace(1,4,Range)
X =[]
x = np.random.rand()*np.ones(Range)
for n in range(Range):
    x = bifurcation(x, r)
    if n > (Range - 200): plt.plot(r, x, 'm.') # restriction for last 200 elements and plotting
    
plt.xlabel('r')
plt.ylabel('x')
plt.show()    
```
